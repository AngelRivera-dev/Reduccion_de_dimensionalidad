```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data  # Características del conjunto de datos
y = iris.target  # Etiquetas de clase

# Aplicar PCA para reducir las dimensiones a 2 componentes
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Crear un DataFrame para facilitar la visualización
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_pca['Clase'] = y  # Añadir la columna de clases

# Visualización de los datos reducidos
plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', hue='Clase', data=df_pca, palette='viridis')
plt.title('PCA aplicado al conjunto de datos Iris')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()

# Mostrar la varianza explicada por cada componente
print(f"Varianza explicada por el primer componente: {pca.explained_variance_ratio_[0] * 100:.2f}%")
print(f"Varianza explicada por el segundo componente: {pca.explained_variance_ratio_[1] * 100:.2f}%")
```
